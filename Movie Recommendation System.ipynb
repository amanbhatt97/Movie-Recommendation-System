#### Since data in 'genres', 'keywords' are in dictionary format, we will convert it into list

```python
movies["genres"][0] #fetching first row of 'genres'
```

```python
import ast #for literal_eval()

#function to convert dictionary value corresponding to 'name' and append them to form a list
def convert(text): 
    L = [] #create an empty list
    for i in ast.literal_eval(text):
        L.append(i['name']) #append all genres which are values corresponding to 'name' into L
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert) #apply convert() to 'genres'
```

```python
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert) #apply convert() to 'keywords'
```

```python
movies
```

```python
movies["cast"][0] #fetching first row of 'cast'
```

#### Since data in 'cast' is in dictionary format, we will convert it into list and fetch first 3 cast

```python
#function to convert dictionary value corresponding to first 3 'name' and append them to form a list
def convert3(text):
    L = [] 
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert) #apply convert() to 'cast'
```

```python
movies.head(2)
```

```python
movies["crew"][0] #fetching first row of 'crew'
```

#### Since data in 'crew' is in dictionary format, we will convert it into list and fetch only director

```python
#function to convert dictionary value corresponding to 'job' = 'Director' and append them to form a list
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director) #apply convert() to 'crew'
```

```python
movies.sample(5) #fetching sample data
```

#### Since data in 'overview' is in string, we will convert it into list

```python
movies['overview'][0] #fetching first row of 'overview'
```

```python
#applyig lambda, split function to convert it into list
movies['overview'] = movies['overview'].apply(lambda x:x.split()) 
```

```python
movies.head(2)
```

#### Removing spaces in names of 'genres', 'keywords', 'cast', 'crew'

```python
movies["genres"] = movies["genres"].apply(lambda x: [i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x: [i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x: [i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.sample(5)
```

#### Create 'tags' column that combine all tags of 'genres', 'keywords', 'cast', 'crew'

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

#### Create a dataframe 'new_df' that contains only columns: 'movie_id', 'title', 'tags'

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

#### Convert 'tags' column from list to string

```python
# using apply, join function to convert list to string
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

#### Performing vectorization by extracting words based on frequency

```python
from sklearn.feature_extraction.text import CountVectorizer
#object 'cv' to perform vectorization by removing stop words and extracting maximum of 5000 features
cv = CountVectorizer(max_features=5000,stop_words='english') 
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() #fitting the vectorization to 'tags' column and convert it into array
```

#### Covert words into their root form to consider similar words as same

```python
import nltk #NLP library
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#function that first converts string to list, then perform stemming and then return as string
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

#### Finding similarities between vectors by performing cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors) #similarity object after performing cosine similarity on vectors
```

```python
similarity[0] #fetching similarity matrix for first row i.e, distances with every movie
```

```python
list(enumerate(similarity[0]))
```

```python
#use enumerate() before using sorted() so that after sorting on basis of similarty, index remain attach
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[1:6]
```

#### Function that recommends 5 similiar movies corresponding to selected one

```python
#find index of selected movie
#go to similarity matrix
#sort movies in descending order of similarity
#fetch top 5 movies
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] #find index of selected movie
    distances = similarity[movie_index] #finding similarities of seleted movie
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6] #sort and fetch top 5 similiar movies
    for i in movie_list: #it will give a tuple
        print(new_df.iloc[i[0]].title) #it will give title
```

```python
recommend("Batman")
```
